## Add annotation data to metadata

```python
# Load annotations
ann = pd.read_csv("./annotations_droplets.csv")
```

```python
ann
```

```python
# Extract cell name
df_filter['cell'] = df_filter['Unnamed: 0'].apply(lambda x: x.replace("-1", ""))
```

```python
df_filter
```

```python
# Merge metadata with annotation data
merg = df_filter.merge(ann, left_on='cell', right_on='cell', how='left')
```

```python
merg.info()
```

```python
# Check if tissues don't match, but this only happens with NaN values
merg[merg['tissue_x'] != merg['tissue_y']]
```

```python
# Remove useless column
merg.drop('tissue_y', axis=1, inplace=True)
merg.rename(columns={'tissue_x': 'tissue'}, inplace=True)
```

```python
merg.info()
```

```python
merg.to_csv("droplet_filter_annotated.csv", index=False)
```

## Create anndata (All tissues)

```python
df_filter_ann = pd.read_csv("./droplet_filter_annotated.csv")
```

```python
all_df = pd.read_csv("./all_tissue_droplet.csv")
```

```python
df_filter_ann.info()
```

```python
df_filter_ann.head()
```

```python
all_df.info()
```

```python
all_df.head()
```

```python
# Drop redundant columns (same value for all rows) and rename initial column
df_filter_ann.drop(columns=['orig.ident', 'percent.ercc'], inplace=True)
df_filter_ann.rename(columns={'Unnamed: 0': 'cell_id'}, inplace=True)

all_df.drop(columns=['orig.ident', 'percent.ercc'], inplace=True)
all_df.rename(columns={'Unnamed: 0': 'cell_id'}, inplace=True)
```

```python
df_filter_ann.head()
```

```python
all_df.head()
```

```python
metadata_cols = all_df.columns
metadata_cols
```

```python
# Read gene counts
raw = mmread("raw_data_all_tissue_droplet.txt")
```

```python
df_raw = pd.DataFrame.sparse.from_spmatrix(raw.T)
```

```python
df_raw
```

```python
# Merge cell metadata with gene conts
merge_df = all_df.join(df_raw)
```

```python
merge_df.head()
```

```python
'''
Filter cells, keep only ones with number of genes per cell greater than 500
and number of UMIs per cell greather than 1000
'''
df_merge_filter = merge_df[(merge_df['nFeature_RNA'] > 500) & (merge_df['nCount_RNA'] > 1000)]
```

```python
df_merge_filter.info()
```

```python
df_merge_filter
```

```python
# Merge metadata + gene counts with annotation data
df_merge_ann = df_filter_ann.merge(df_merge_filter, left_on='cell_id', right_on='cell_id', how='right')
```

```python
df_merge_ann.info()
```

```python
df_merge_ann.columns[0:30]
```

```python
df_merge_ann[['nCount_RNA_x', 'nFeature_RNA_x', 'channel_x', 'mouse.id_x', 'tissue_x', 'subtissue_x', 'mouse.sex_x', 'percent.ribo_x', 'percent.Rn45s_x',
                         'cell', 'cell_ontology_class', 'cell_ontology_term_iri', 'cell_ontology_id']]
```

```python
# Drop redundant columns
df_merge_ann.drop(columns=['nCount_RNA_y', 'nFeature_RNA_y', 'channel_y', 'mouse.id_y', 'tissue_y', 'subtissue_y', 'mouse.sex_y', 'percent.ribo_y', 'percent.Rn45s_y'], inplace=True)
```

```python
df_merge_ann.columns[0:30]
```

```python
# Rename columns
df_merge_ann.rename(columns={'nCount_RNA_x':'nCount_RNA', 'nFeature_RNA_x':'nFeature_RNA', 'channel_x':'channel',
                   'mouse.id_x':'mouse.id', 'tissue_x':'tissue', 'subtissue_x':'subtissue', 'mouse.sex_x':'mouse.sex',
               'percent.ribo_x':'percent.ribo', 'percent.Rn45s_x':'percent.Rn45s'}, inplace=True)
```

```python
df_merge_ann.columns[0:30]
```

```python
df_merge_ann.head()
```

```python
# Save files
df_merge_ann.to_csv("ann_tissues_filtered.csv", index=False)
df_merge_filter.to_csv("all_tissue_droplet_filter_raw.csv", sep="#", index=False)
```
